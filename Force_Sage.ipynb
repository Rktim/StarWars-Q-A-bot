```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
burl="https://starwars.fandom.com/wiki/"
```

```python
topics=['Luke','Leia','Han','Chewbacca','Darth Vader','Yoda',"Obi-Wan Kenoboi"]
```

```python
def scrape(topics):
  url=burl+topics
  response=requests.get(url)
  if response.status_code==200:
    soup=BeautifulSoup(response.content,'html.parser')
    title=soup.find('h1',{'id':'firstHeading'}).text.strip()
    content=soup.find('div',{'class':'mw-parser-output'}).text.strip()
    return {"title":title,"content":content}
  else:
    return None
```

```python
data=[]
for topic in topics:
  a=scrape(topic)
  if a :
    data.append(a)
```

```python
df=pd.DataFrame(data)
df.to_csv('starwars.csv')
```

Setting Up The ChatBot.

```python
import cohere
import langchain
from langchain import LLMChain ,Cohere
from langchain.prompts.chat import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate
from langchain.schema import HumanMessage, SystemMessage, AIMessage
```

```python
from google.colab import userdata
api_key = userdata.get('cohere')
```

```python
prmpt = ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template("Hi, I'm a Star Wars Expert."),
    HumanMessagePromptTemplate.from_template("Answer the following question: {input}")
])
```

```python
llm = Cohere(cohere_api_key=api_key)
```

```python
starwars_chain = LLMChain(llm=llm, prompt=prmpt)
```

```python
q="Who is Darth Vader?"
res=starwars_chain.run(q)
print(res)
```

Chat Bot Interface .

```python
def chat_with_bot():
    print("Welcome to the Star Wars Q&A Chatbot!")
    print("Type 'exit' to end the conversation.\n")

    while True:
        question = input("You: ")
        if question.lower() == 'exit':
            print("May The Force Be With You!")
            break

        response = starwars_chain.run({"input": question})
        print(f"Force Sage: {response}")


if __name__ == "__main__":
    chat_with_bot()
```
